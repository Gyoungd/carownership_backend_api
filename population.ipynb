## Building DataBase for Population in Melbourne

**Reference**

Australian Bureau of Statistics. (2022, July 26). *Regional population, 2021: Population estimates by LGA and electoral division (ASGS2021), 2001 to 2021 – Final* [Data set]. https://www.abs.gov.au/statistics/people/population/regional-population/2021

```python
import pandas as pd
```

```python
# Load the data sheet 1(Table 1)
df = pd.read_excel('./ABS Raw Datasets/ABS_Population_Estimates.xlsx', sheet_name='Table 1')

# Display the first few rows of the dtaaset
df.head(10)
```

```python
# Drop the first 5 rows
df = df.drop(range(5))

# Reset the index
df = df.reset_index(drop=True)

# Display the first few rows of the dataset
df.head(10)
```

```python
df.iloc[0:2,10:]

# Drop column 32-34
df = df.drop(columns=df.columns[31:35])

df.iloc[0:2,10:]
```

```python
# Get the list of header column 10-
header_col1 = df.iloc[0, 10:]

# Get the list of header column 0-26
header_col2 = df.iloc[1, 0:10]

# Merge the two lists
header_col = pd.concat([header_col2, header_col1]).tolist()

header_col
```

```python
# Change the header names, same as header_col

current_columns = df.columns.tolist()

# Mapping the column names
rename_dict = dict(zip(current_columns, header_col))

# rename()
df.rename(columns=rename_dict, inplace=True)

# Drop the first 2 rows
df = df.drop(range(3))
df = df.reset_index(drop=True)

# Check the first few rows of the dataset
df.head()
```

```python

# Filter out the data using SA4 name contains 'Melbourne'
df_melbourne = df[df['SA4 name'].str.contains('Melbourne', case=False, na=False)]
df_melbourne = df_melbourne.reset_index(drop=True)

df_melbourne.head()
```

```python
df_melbourne.info()
```

```python
# Filter out Melbourne CBD using SA2 name contains 'Melbourne CBD'
df_CBD = df_melbourne[df_melbourne['SA2 name'].str.contains('Melbourne CBD', case=False, na=False)]
df_CBD = df_CBD.reset_index(drop=True)

df_CBD.head()
```

```python
df_CBD.info()
```

### Table Structure Example

**Population**

| population_id | region_name | region_code | year | population |
|:---|:---|:---|:---|:---|
|1|Melbourne CBD - East | 206041503	| 2014 | 10455 |

```python
# column names
df_CBD.columns
```

```python
# Remain SA2 name, SA2 code, years(2001-2021)
df_CBD = df_CBD[['SA2 name', 'SA2 code', 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]]

# Rename columns
df_CBD.rename(columns = {'SA2 name': 'region_name', 'SA2 code': 'region_code'}, inplace = True)

# Get the year list
year_list = df_CBD.columns[2:].tolist()
print(year_list)

# Pivot the table: region_name, region_code, year, population
df_CBD = df_CBD.melt(id_vars=['region_name', 'region_code'],
                     value_vars=year_list,
                     var_name='year',
                     value_name='population')

# Check the table
print(df_CBD.head())

# Check the table structure
print(df_CBD.info())
```

```python
# Convert Datatype
df_CBD['region_code'] = df_CBD['region_code'].astype(int)
df_CBD['year'] = df_CBD['year'].astype(int)
df_CBD['population'] = df_CBD['population'].astype(int)

# Check the table structure
print(df_CBD.info())
```

```python
df_CBD.head()
```

```python
# Calculate Melbourne CBD population yearly by summing up the population of East, North, West
total_rows = []

for year in df_CBD['year'].unique():
    # Filter by years
    pop_year = df_CBD[df_CBD['year'] == year]

    # Calculate the total population
    total_pop = pop_year['population'].sum()

    # Create a new row dictionary (region_code = None)
    new_row = {
        'region_name' : 'Melbourne CBD - Total',
        'region_code' : None, # None because it is not a real region code
        'year' : year,
        'population' : total_pop
    }

    total_rows.append(new_row)

df_total_rows = pd.DataFrame(total_rows)

# Append total rows to the original table
df_CBD = pd.concat([df_CBD, df_total_rows], ignore_index=True)

# Add population_id column
df_CBD['population_id'] = range(1, len(df_CBD) + 1)

df_CBD = df_CBD[['population_id', 'region_name', 'region_code', 'year', 'population']]

# Check the table
df_CBD.tail()

```

```python
df_CBD.info()
```

```python
# Check the maximum length of region_code and region_name
max_region_name_length = df_CBD['region_name'].apply(len).max()
max_region_code_length = df_CBD['region_code'].apply(lambda x: len(str(x)) if x is not None else 0).max()

print(f"Maximum length of region_name: {max_region_name_length}")
print(f"Maximum length of region_code: {max_region_code_length}")
```
