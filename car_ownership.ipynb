## Building DataBase for Car Ownership Analysis

```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
path = 'ABS Raw Datasets'
df_2016 = pd.read_excel(path +'/car2016.xlsx')
df_2017 = pd.read_excel(path +'/car2017.xlsx')
df_2018 = pd.read_excel(path +'/car2018.xlsx')
df_2019 = pd.read_excel(path +'/car2019.xlsx')
df_2020 = pd.read_excel(path +'/car2020.xlsx')
df_2021 = pd.read_excel(path +'/car2021.xlsx')


df_2016
```

```python
# Cleaning the dataset
# Drop the row 0-6
df_2016 = df_2016.drop(index=range(0, 7))
df_2017 = df_2017.drop(index=range(0, 7))
df_2018 = df_2018.drop(index=range(0, 7))
df_2019 = df_2019.drop(index=range(0, 7))
df_2020 = df_2020.drop(index=range(0, 7))
df_2021 = df_2021.drop(index=range(0, 7))

# Drop the first column and the third column
df_2016 = df_2016.drop(columns=df_2016.columns[0])
df_2017 = df_2017.drop(columns=df_2017.columns[0])
df_2018 = df_2018.drop(columns=df_2018.columns[0])
df_2019 = df_2019.drop(columns=df_2019.columns[0])
df_2020 = df_2020.drop(columns=df_2020.columns[0])
df_2021 = df_2021.drop(columns=df_2021.columns[0])

# Change the column names
df_columns = ['State', 'Total_cars', 'Drop']

df_2016.columns = df_columns
df_2017.columns = df_columns
df_2018.columns = df_columns
df_2019.columns = df_columns
df_2020.columns = df_columns
df_2021.columns = df_columns

# Drop the first two rows
df_2016 = df_2016.iloc[2:,:]
df_2017 = df_2017.iloc[2:,:]
df_2018 = df_2018.iloc[2:,:]
df_2019 = df_2019.iloc[2:,:]
df_2020 = df_2020.iloc[2:,:]
df_2021 = df_2021.iloc[2:,:]

# Drop the 'Drop' column
df_2016 = df_2016.drop(columns=['Drop'])
df_2017 = df_2017.drop(columns=['Drop'])
df_2018 = df_2018.drop(columns=['Drop'])
df_2019 = df_2019.drop(columns=['Drop'])
df_2020 = df_2020.drop(columns=['Drop'])
df_2021 = df_2021.drop(columns=['Drop'])

df_2016
```

```python
# Order by 'State' column
df_2016 = df_2016.sort_values(by='State')
df_2017 = df_2017.sort_values(by='State')
df_2018 = df_2018.sort_values(by='State')
df_2019 = df_2019.sort_values(by='State')
df_2020 = df_2020.sort_values(by='State')
df_2021 = df_2021.sort_values(by='State')

# Remove the missing values
df_2016 = df_2016.dropna()
df_2017 = df_2017.dropna()
df_2018 = df_2018.dropna()
df_2019 = df_2019.dropna()
df_2020 = df_2020.dropna()
df_2021 = df_2021.dropna()

# Reset the index
df_2016 = df_2016.reset_index(drop=True)
df_2017 = df_2017.reset_index(drop=True)
df_2018 = df_2018.reset_index(drop=True)
df_2019 = df_2019.reset_index(drop=True)
df_2020 = df_2020.reset_index(drop=True)
df_2021 = df_2021.reset_index(drop=True)

# Convert 'Total_cars' to numeric
df_2016['Total_cars'] = pd.to_numeric(df_2016['Total_cars'])
df_2017['Total_cars'] = pd.to_numeric(df_2017['Total_cars'])
df_2018['Total_cars'] = pd.to_numeric(df_2018['Total_cars'])
df_2019['Total_cars'] = pd.to_numeric(df_2019['Total_cars'])
df_2020['Total_cars'] = pd.to_numeric(df_2020['Total_cars'])
df_2021['Total_cars'] = pd.to_numeric(df_2021['Total_cars'])

df_2016
```

```python
# Add a new column 'Year' to each dataframe
df_2016['Year'] = 2016
df_2017['Year'] = 2017
df_2018['Year'] = 2018
df_2019['Year'] = 2019
df_2020['Year'] = 2020
df_2021['Year'] = 2021

# Merge all dataframes into one
df_all = pd.concat([df_2016, df_2017, df_2018, df_2019, df_2020, df_2021], ignore_index=True)

# Remove the row with 'Total' in 'State' column
df_all = df_all[df_all['State'] != 'Total']

# Reset the index of the merged dataframe
df_all = df_all.reset_index(drop=True)

df_all
```

```python
df_all['State'].unique()
```

```python
# Change the State naems to abbreviations
state_abbreviations = {
    'AUSTRALIAN CAPITAL TERRITORY': 'ACT',
    'NEW SOUTH WALES': 'NSW',
    'NORTHERN TERRITORY': 'NT',
    'QUEENSLAND': 'QLD',
    'SOUTH AUSTRALIA': 'SA',
    'TASMANIA': 'TAS',
    'VICTORIA': 'VIC',
    'WESTERN AUSTRALIA': 'WA'
}
df_all['State'] = df_all['State'].replace(state_abbreviations)

# Add columns 'carowner_id'
df_all['Carowner_id'] = df_all.index + 1

# Reorder the columns
df_all = df_all[['Carowner_id', 'State', 'Year', 'Total_cars']]

df_all
```

```python
# SAVE THE CLEANED DATAFRAME
df_all.to_csv('car_ownership_cleaned.csv', index=False)
```

```python
# Check the trend of car ownership in Victoria
df_vic = df_all[df_all['State'] == 'VIC']

df_vic

# Plotting the trend of car ownership in Victoria
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_vic, x='Year', y='Total_cars', marker='o')
plt.title('Trend of Car Ownership in Melbourne (2016-2021)')
plt.xlabel('Year')
plt.ylabel('Total Cars')
min_cars = df_vic['Total_cars'].min()
max_cars = df_vic['Total_cars'].max()

n_ticks = 7 
step = int(np.ceil((max_cars - min_cars) / (n_ticks - 1) / 10000) * 10000)
start = int(np.floor(min_cars / step) * step)
end = int(np.ceil(max_cars / step) * step)
yticks = np.arange(start, end + step, step)

#yticklabels = [f'{int(y):,}' for y in yticks]  # f'{y/1e6:.1f}M'
yticklabels = [f'{y/1e6:.1f}M' for y in yticks]

plt.yticks(yticks, yticklabels)

plt.xticks(df_vic['Year'], rotation=45)
plt.grid()
plt.show()

```
